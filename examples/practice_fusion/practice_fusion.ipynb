```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://rxnav.nlm.nih.gov/REST/Ndfrt/search'
url_info = 'https://rxnav.nlm.nih.gov/REST/Ndfrt/allInfo.json'
```

```python
result = requests.request('GET', url, params={'conceptName': 'Lisinopril', 'kindName': 'DRUG_KIND'})
```

```python
result.content
```

```python
soup = BeautifulSoup(result.content, "lxml-xml")
```

```python
soup
```

```python
nui = soup.find('conceptNui')
```

```python
nui
```

```python
nui.string
```

```python
nui_info = requests.request('GET', url_info, params={'nui': nui.string})
```
